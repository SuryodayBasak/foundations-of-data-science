```python
#Author: Suryoday Basak
#suryodaybasak.info
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
mpl.rcParams['figure.figsize'] = (10,8)
```

```python
#Reading the data
df = pd.read_csv('../datsets/cars/mtcars.csv')
print(df)
```

```python
data_mat = df.values
r,c = np.shape(data_mat)
print(np.shape(data_mat))
print(r)
ones_stub = np.ones((r,))
data_mat = np.c_[data_mat, ones_stub]
print(data_mat)
```

```python
#Define x and y
x = data_mat[:,2:]
y = data_mat[:,1]

print('The x is:')
print(x)
print('')
print('The y is:')
print(y)
```

```python
#Visualize cyl vs mpg
plt.scatter(x[:,0], y)
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
#Visualize hp vs mpg
plt.scatter(x[:,2], y)
plt.xlabel('Number of Cylinders')
plt.ylabel('HP')
plt.show()
```

```python
#Find the number of features
_,n = np.shape(x)
print(n)
```

```python
#Our matrix system will have n+1 entries, with the 1 addition corresponding to the intercept
A = np.zeros((n,n))
print(A)
```

```python
for i in range(0,n):
    for j in range(0,n):
        A[i,j] = np.sum(np.multiply(x[:,i],x[:,j]))

#Populating the last row of the matrix
for j in range(0,n):
    A[n-1,j] = np.sum(sum(x[:,j]))

print(A)
```

```python
#This vector will store the RHS of the matrix system
b = np.zeros((n,))

for i in range(0,n):
    b[i] = np.sum(x[:,i]*y)
#b[n] = np.sum(y)                #This term will hold sum(yi)

print(b)
```

```python
#Finding the inverse
A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
#Computing the slopes corresponding to each variable
theta = np.matmul(A_inv,b)
print(theta)
```

```python
#Finding the error and displaying the difference between original and estimated
error = 0.0
r, c = np.shape(x)
for i in range(0,r):
    i_est = np.sum(x[i,:]*theta)
    error += (i_est - y[i])**2 #(Estimated - original)^2
    print('Original:\t', y[i], '\tEstimated:\t',i_est)
          
error/=n
print("The mean squared error is:\t\t", error)
print("The root means squared error is:\t", error**(0.5))
```
